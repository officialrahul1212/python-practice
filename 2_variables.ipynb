# Variables Written Rules
1) 123abc ❌   = can't start with numbers
2) abc123 ✅   = you can use numbers in end
3) abc5xyz ✅  = use no in middle
4) abc$xyz  ❌ = can't use special characters /symbols
5) _ abc_ bcd _ ✅ =use underscore
6) if        ❌ = not use reserved keywords

_reserved keywords=
if, else, def, print, ifelse, true, false, id, and, or, etc

```python
MH10='sangli'
print(MH10)

MH_09="kholapur"
print(MH_09)

MH10Sangli="Sangli City"
print(MH10Sangli)
```

# Data Types
Data types are the types of data that can be stored in a variable
(Same type of data store in variable)

1] Numbers -> int, float, complex

```python
a= 10            # is an integer (number)
print(a)
print(type(a))

b=10.5           # is an float(Decimal)
print(b)
print(type(b))

c=3+0j           # is an complex
print(c)
print(type(c))
```

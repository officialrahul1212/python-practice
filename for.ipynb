### For Loop

The for loop in Python is used for iteration, allowing us to execute a block of code once for each element in a sequence.

A for loop is used to repeat a task for each item in a collection.

👉 Used to iterate over a sequence like:

list
string
tuple
range()

Syntax

for variable in sequence:
    statement


range() Function in for Loop


📌 Syntax
range(start, stop, step)

```python
for i in range(6):      #range(start, stop, step)    0 , 5
    print(i)
```

```python
for i in range(1,6):   #range(start, stop, step)      (1,5)
    print(i)
```

```python
for i in range(1,11,3):  #range(start, stop, step)  ( 1,11,2)
    print(i)
```

```python
for i in range(11,1,-2):  #range(start, stop, step)  (11,1,-2)
    print(i)
```

```python
fruits = ["apple", "banana", "mango"]

for i in fruits:
    print(i)
```

```python
name = "Python"

for z in name:
    print(z)
```

```python
num = 5

for i in range(1, 11):
    #print(num, "x", i, "=", num * i)
    print(f'{num} x {i} = {num*i}')


# n = 5
# for i in range(1, 11):
#     print(n * i)
```

### Loop Control Statement

1.break statement <br>
2.continue statement<br>
3.pass statement

```python
# 1. break Statement
# 👉 break immediately stops the loop completely.
# 👉 Once break executes, the loop terminates.

for i in range(1, 6):

    if i == 4: #4==4
        break

    print(i)   #3
```

```python
# 2. continue Statement
# 👉 continue skips the current iteration and moves to the next iteration.
# 👉 Loop does NOT stop.

for i in range(1, 6):

    if i == 3:    # 4==3
        continue

    print(i)     #4
```

```python
# 3. pass Statement

# 👉 pass is a placeholder statement.
# 👉 It does nothing.
# 👉 Used when code will be added later.


#     Why pass is Used?

# ✅ Avoids syntax error
# ✅ Used during future planning
# ✅ Creates empty loops/functions/classes temporarily


for i in range(5):
    pass
print(i)
```

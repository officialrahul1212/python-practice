### Tuple ( )
A tuple is a ordered collection of datatype/elements  used to store multiple values in single variaable

A tuple is an immutable, ordered collection. Once created, its elements cannot be changed. Tuples are
faster and more memory-efficient than lists.

list store values in ()

### Characteristics
- Ordered — maintains insertion order
- Immutable — cannot be modified after creation
- Allows duplicates
- faster than list - use less-memory
- Store different datatypes
- supporting indexing & slicing

### Create Tuple

```python
T=()                 #empty Tuple
print("empty Tuple = ",T)

T1=(1,2,3,4,5)
print("Tuple =",T1)

T2=(1,23.5,'abc',3)               #mixed datatype Tuple
print("mixed datatype Tuple =",T2)

T3=tuple(range(1,6))
print("using range function create Tuple =",T3)

T4=tuple([1,2,3,4])
print("convert Tuple into tuple=",T4)
```

### 🔶Tuple Methods  (only 2 — immutable)
count(), index()


```python
a=(1,2,3,4,3,6)
print(a)


#1️⃣count() -count if any value how many times in the tuple
print("count of value=",a.count(3))

#2️⃣index() -returen index position of value
print("index position is=",a.index(4))
```

### 🔷Indexing & Slicing

- _positive indesing start with 0 (0 1 2 3 4)_
- _negative indexing start with -1(-5 -4 -3 -2 -1) access elements last to 1st_

```python
num=(1,2,3,5,6,7,8)

#1️⃣indexing -access single element using indexing
print("sing index no to acces element=",num[1])     #using index no to acces element
print("access last emement using negative indexing=",num[-1])    # access last emement using negative indexing

print(num[2])
print(num[-2])

#2️⃣Slicing access multiple emements in the tuple usind index
#syntax =    print(tuple(start,end,step))

print(num[1:4])    # access 1 to 3 values  [1,2,3]
print(num[:4])    # acces 4 index values  [1,2,3,5]
print(num[2:])
print(num[::])
print(num[::-2])
         
```

### 🔷Tuple Operations

```python
n1=(1,2)
n2=(3,4)

#1️⃣concatenation(+) = combine 2 or more tuple to generate a new tuple
print(n1+n1)

#2️⃣repetation(*) =deplicates the element of q tuple a specific no of times
print(n1*3)

#3️⃣membership (in,not in) =check the whether item present/exist in tuple or not 
# and return boolean value

color=('red','black','pink')
print('red' in color)
print('red' not in color)

#4️⃣iteration = one by on values print
#print(for i in n1:)
```

### Tuple Built-in Functions

```python
no=(0,1,3,2,5,6)

#1️⃣.len() -return total no of itemps present in the tuple (count the total no.s in tuple)
print("len()   =") 
print(len(no))

#2️⃣.type() -return data type
print("\n type()  =")
print(type(no))

#3️⃣.max() -return largest value in the tuple
print("max()  =")
print(max(no))

#4️⃣min() -return smallest value in the tuple
print("min()  =")
print(min(no))

#5️⃣sum()- return the total sum of the tuple
print("sum()  =")
print(sum(no))

#6️⃣sorted() - return sorted new tuple (alpahbetically)
print("sorted() =")
print(sorted(no)) 
```

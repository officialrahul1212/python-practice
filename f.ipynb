Q24. find 2nd largest number

```python
num=[1,2,3,4,6,7,5,9]
largest=0
second_larg=0

for i in num:
    if i>largest:
        second_larg=largest
        largest=i
    elif i<largest and i>second_larg:
        second_larg=i

print(second_larg)
```

Q27. reverse a number

```python
num=12345
digit=0
rev=0
while num > 0 :
    digit=num % 10                 #get/ give last digit
    rev=rev * 10 + digit
    num=num // 10                   #cut/delete last digit
print(rev) 
```

Q28. sum of digit

```python
num=12345
digit=0
total=0
while num > 0 :
    digit=num % 10
    total=total + digit
    num=num // 10
print(total) 
```

Q29. check palindrom

```python
num=121
temp=num
digit=0
p=0

while num > 0:
    digit=num%10
    p=p*10+digit
    num=num//10

if p==temp:
    print(f'{temp} is palindrom')
else:
    print(f'{temp} is not palindrom')
```

Q30 Armstrong number

```python
num=153
temp=num
digit=0
total=0

while num>0:
    digit=num % 10
    total=total + digit **3
    num=num // 10

if total==temp:
    print(f'{temp} is armstrong')
else:
    print(f'{temp} is not armstrong')
```

### List

A list is a ordered collection of datatype used to store multiple values in a single variable.

A list is a mutable, that allows duplicate elements. It is the most commonly used data structure in Python.

list store values in []

### Characteristics
- Ordered — maintains insertion order
- Mutable — can change after creation
- Allows duplicates
- Supports any data type (heterogeneous)
- Indexed from 0 (also supports negative indexing)

### Create List

```python
l=[]                 #empty list
print("empty list = ",l)

l1=[1,2,3,4,5]
print("list =",l1)

l2=[1,23.5,'abc',3]               #mixed datatype list
print("mixed datatype list =",l2)

l3=list(range(1,6))
print("using range function create list =",l3)

l4 = [x**2 for x in range(6)]        # list comprehension
print("list comprehension =",l4)
```

### 🔶List Methods


- _🔸Add methods_

```python
#1️⃣.append() -Add element/item at end of the list (only 1 element add in end)

a=[1,2,3,4,5]
print(a)

a.append(6)                                  #append/ add item 6 ate end of the list
print("after append 6 item in list = ",a)


#2️⃣.extend() -Add multiple element/item at end of the list (multiple element add in end)

a.extend([7,8,9])                                #ectend/ add item 7,8,9 at end of the list
print("after extend add 7,8,9 at end if list =",a)


#3️⃣.insert() - Add element/item at specific location

a.insert(4,'five')
print("after insert add five in the 4th location =",a)
```

- _🔹Remove/deletion Methods_

```python
b=[1,2,4,5,6,777,7,8,9,10]
print(b)

#4️⃣.remove() - remove used to  remove specific value

b.remove(777)          
print("remove use to remove index value =",b)


#5️⃣.pop()  - remove last element if not give any index value 
                     #if give index  value than remove last element
b.pop()
print("pop use to value =",b)


#6️⃣.clear()  - remove all elements in list (return blank list)

b.clear()
print("clear use to remove all value =",b)
```

```python
c=[1,4,5,30,5,2,3]
print(c)

#7️⃣.index() - find the index position
d=c.index(30)
print("index value of 30 is =",d)


#8️⃣.count() - count the specific value/item how many time (repeate) in the list
e=c.count(5)            # 5 value 2 times in the list
print("count value how many come in list =",e)


#9️⃣sort()   - sort the original list (ascending order)
c.sort()
print("sort the original list =",c)
#c.sort(reverse=True)                     # reverse sort using sort function
#print(c)

#🔟reverse() -reverse the list (descending order)
c.reverse()
print("revrse the list=",c)


#1️⃣1️⃣.copy()  - copy the list and create another list
f=c.copy()
print("create another copy of the list",f)
```

### 🔷Indexing & Slicing

- _positive indesing start with 0 (0 1 2 3 4)_
- _negative indexing start with -1(-5 -4 -3 -2 -1) access elements last to 1st_

```python
num=[1,2,3,5,6,7,8]

#1️⃣indexing -access single element using indexing
print("sing index no to acces element=",num[1])     #using index no to acces element
print("access last emement using negative indexing=",num[-1])    # access last emement using negative indexing

print(num[2])
print(num[-2])

#2️⃣Slicing access multiple emements in the list usind index
#syntax =    print(list(start,end,step))

print(num[1:4])    # access 1 to 3 values  [1,2,3]
print(num[:4])    # acces 4 index values  [1,2,3,5]
print(num[2:])
print(num[::])
print(num[::-2])
```

### 🔶List Built-in Functions

```python
no=[0,1,3,2,5,6]

#1️⃣.len() -return total no of itemps present in the list (count the total no.s in list)
print("len()   =") 
print(len(no))

#2️⃣.type() -return data type
print("\n type()  =")
print(type(no))

#3️⃣.max() -return largest value in the list
print("max()  =")
print(max(no))

#4️⃣min() -return smallest value in the list
print("min()  =")
print(min(no))

#5️⃣sum()- return the total sum of the list
print("sum()  =")
print(sum(no))

#6️⃣sorted() - return sorted new list (alpahbetically)
print("sorted() =")
print(sorted(no)) 

#7️⃣list() -convert in list (type coversion)
x="abc"
print("list()  =")
print(list(x))

#8️⃣any() - return true if any value is true
print("any()  =")
print(any(no))

#9️⃣all() -return true if all values are true
print("all()  =")
print(all(no))

#🔟enumerate() - add index with values
print("enumerate()  =")
print(enumerate(no))
```

### While Loop

A while loop is used to execute a block of code repeatedly as long as a condition is true.
- It is called a condition-based loop.

Some easy-to-remember versions:

- "A while loop repeats a block of code until a condition becomes false."
- "A while loop runs as long as a specified condition is true."
- "A while loop is used when the number of iterations is not known in advance." 
-  Good interview answer.
- "A while loop continuously executes code based on a condition."


Remember:

🔸for loop → Number of iterations usually known.<br>
🔹while loop → Number of iterations usually unknown; depends on a condition.

Flow
- Check the condition.
- If True, execute the code block.
- Again check the condition.
- Repeat until the condition becomes False.



syntax= 
while condition:
    # code block

```python
x = 1

while x <= 5:   #6   = false
    print(x)   #1,2 ,3, 4, 5
    x += 1    #  x=x+1 = 6
```

```python
x = 5

while x >= 1:  # 0
    print(x)   #1
    x -= 1     # x=x-1
```

```python
password = ""

while password != "python":
    password = input("Enter password: ")

print("Access Granted")
```

```python
i = 1
total = 0

while i <= 10:  # 11  false
    total += i  #total=total+i   =45+10=55    
    i += 1      #i=i+1  # 11

print(total)
```

```python
num = 5
i = 1

while i <= 10:
    print(num, "x", i, "=", num * i)
    i += 1
```

```python
# while with else

# The else block runs when the loop ends normally.

i = 1
while i <= 3:
    print(i)
    i += 1
else:
    print("Loop Finished")
```

### Loop Control Statement

1.break statement <br>
2.continue statement<br>
3.pass statement

```python
# 1. break Statement
# 👉 break immediately stops the loop completely.
# 👉 Once break executes, the loop terminates.

x = 1

while x <= 5:

    if x == 4:
        break

    print(x)
    x += 1
```

```python
# 2. continue Statement
# 👉 continue skips the current iteration and moves to the next iteration.
# 👉 Loop does NOT stop.

x = 0
while x < 5:
    x += 1
    if x == 3:
        continue

    print(x)
```

```python
# 3. pass Statement

# 👉 pass is a placeholder statement.
# 👉 It does nothing.
# 👉 Used when code will be added later.


#     Why pass is Used?

# ✅ Avoids syntax error
# ✅ Used during future planning
# ✅ Creates empty loops/functions/classes temporarily


x = 1

while x <= 5:
    pass
```

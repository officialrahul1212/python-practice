# Most Common Interview Programs on for and while Loops

 ## 1. Print Numbers from 1 to 10 

```python
for i in range(1, 11):
    print(i)
```

```python
i = 1

while i <= 10:
    print(i)
    i += 1
```

## 3. Sum of First 10 Numbers

```python
total = 0

for i in range(1, 11):
    total += i

print(total)
```

```python
i = 1
total = 0

while i <= 10:
    total += i
    i += 1

print(total)
```

## 4. Multiplication Table

```python
num = 5

for i in range(1, 11):
    #print(num, "x", i, "=", num * i)
    print(f'{num} x {i} = {num*i}')
```

```python
num = 5
i = 1

while i <= 10:
    #print(num, "x", i, "=", num * i)
    print(f'{num} x {i} = {num*i}')

    i += 1
```

## 5. Factorial Program
factorial of 5 = 5 x 4 x 3 x 2 x 1

```python
num = 5
fact = 1

for i in range(1, num + 1): #(1,6)
    fact *= i    # fact = fact*i    #1*1 = 1

print(fact)
```

```python
num = 5
fact = 1
i = 1

while i <= num:
    fact *= i      # fact = fact * i
    i += 1
print(fact)
```

## 6. Reverse Number

```python
num = int(input("Enter number: "))
reverse = 0

while num > 0:       # 0 > 0 

    digit = num % 10                         # digit = 539 % 10 = 9
    reverse = reverse * 10 + digit           # reverse = 0 * 10 + 9 = 5
    num = num // 10                          # num = 5 // 10 = 0

print("Reverse Number =", reverse)           # 8935


#5398   -- >   8935
```

```python
num = 12345
rev = 0

for i in str(num):
    rev = int(i) + rev * 10

print(rev)
```

## 7. Palindrome Number
abc = cba   or   abcd = dcba    then its palindrome no.

```python
num = 1219
temp = num
reverse = 0

while num > 0:

    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if temp == reverse:
    print("Palindrome")

else:
    print("Not Palindrome")
```

## 8. Fibonacci Series
a, b , a+b , b+(a+b), .....

```python
a = 1
b = 2

for i in range(10):

    print(a)   # 1 ,2 ,3,

    c = a + b    # 2+3 = 5
    a = b        #a=3
    b = c        #b=5
```

## 9. Count Digits in Numbe

```python
num = int(input("Enter number: "))

count = 0

while num > 0:    # 0 >0 

    count += 1             #4
    num = num // 10        #num = 6 // 10 = 0

print("Total Digits =", count)
```

## 10. Armstrong Number

```python
Sum of cube of each digit = original number

no = 153

= 1³ + 5³ + 3³ 
= 1 + 125 + 27
=153
```

```python
num = 153
temp = num
sum = 0

while num > 0:    # 0 >0

    digit = num % 10            # = 1 % 10 = 1
    sum = sum + ( digit ** 3 )   # 152 + (1 ** 3) = 153
    num = num // 10             # 1 // 10 = 0

if temp == sum:
    print("Armstrong Number")

else:
    print("Not Armstrong")
```

## Prime number

## Largest number 

```python
# List of numbers जिनमें से maximum ढूंढना है
nums = [10, 45, 23, 99, 12, 100]

# 'a' variable में अभी तक का सबसे बड़ा number store होगा
# शुरू में 0 रखते हैं (assume: सारे numbers 0 से बड़े हैं)
a = 0

# List के हर number पर एक-एक करके जाओ
for i in nums:

    # Check करो: क्या current number (i) अब तक के max (a) से बड़ा है?
    if i > a:

        # अगर हाँ, तो नया maximum save कर लो
        a = i

# Loop खत्म — 'a' में अब सबसे बड़ा number है
print(a)  # Output: 100
```

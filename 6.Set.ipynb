### Set {}

A set is a unordered collection of unique elements to store multiple values in a single variable.

set does not allow duplicate values

list store values in {}

### Characteristics
- Unordered — no guaranteed order
- Mutable — elements can be added/removed
- No duplicates — automatically removes them
- Heterogeneous data allowed (Multiple data types)
- No indexing — cannot use set[0] =(because set is unordered)
- Elements must be hashable (no lists inside a set)

### 🔴Create Set

```python
s={}                 #empty set
print("empty set = ",s)

s1={1,2,3,4,5}
print("set =",s1)

s2={1,23.5,'abc',3}               #mixed datatype set
print("mixed datatype set =",s2)

s3=set(range(1,6))
print("using range function create set =",s3)

s4 = set('hello')            # {'h','e','l','o'}
print(s4)

s5 = {x**2 for x in range(6)}        # set comprehension
print("set comprehension =",s5)
```

### 🔷Set Methods


```python
a={1,2,3,4,5}
print(a)

#1️⃣.add() -Add one element/item 
a.add(6)                                  #add item in set (unordered)
print("add item in set = ",a)


#2️⃣.update() -Add multiple element/item 

a.update([7,8,9])                                #ectend/ add multiple item 7,8,9 at set
print("after update add 7,8,9 in set =",a)


#3️⃣.remove() - remove specific item
a.remove(8)
print("remove specific item=",a)     #error if specific value not found


#4️⃣.discard() - remove specific value without error

a.discard(777)          # specific value not found in set than not give the eror
print("remove specific value without error =",a)


#5️⃣.pop()  - remove random element 
                
a.pop()
print("pop use to value =",a)


#6️⃣.clear()  - remove all elements in set (return blank set)

a.clear()
print("clear use to remove all value =",a)


#7️⃣copy() - create copy of set
b=a.copy()
print("copy the set=",b)
```

### 🟩Set Operations

```python
x={1,2,3,4}
y={3,4,5,6}

#1️⃣union() - combine all unique elements
print(x.union(y))                         #o/p= {1, 2, 3, 4, 5, 6}

#2️⃣intersection() -return common elements
print(x.intersection(y))                  #o/p= {3, 4}

#3️⃣difference() -  non common element present in 1st set only
print(x.difference(y))                     #o/p= {1, 2}

#4️⃣symmetric_difference() -return non common elements in both set
print(x.symmetric_difference(y))          #o/p= {1, 2, 5, 6}

#5️⃣membership_operator -in, not in
s1={10,20,30,40}
print(20 in s1)                           #o/p= True  
print(20 not in s1)                       #o/p= False

```

### 🔶Set Built-in Functions

```python
no={0,1,3,2,5,6}

#1️⃣.len() -return total no of itemps present in the set (count the total no.s in set)
print("len()   =") 
print(len(no))

#2️⃣.type() -return data type
print("\n type()  =")
print(type(no))

#3️⃣.max() -return largest value in the set
print("max()  =")
print(max(no))

#4️⃣min() -return smallest value in the set
print("min()  =")
print(min(no))

#5️⃣sum()- return the total sum of the set
print("sum()  =")
print(sum(no))

#6️⃣sorted() - return sorted new set (alpahbetically)
print("sorted() =")
print(sorted(no)) 


```

# 🟫What is a FrozenSet?

A FrozenSet is an immutable version of a set.

characteristics:

un_Ordered
Im_Mutable (un_changeable)
Un_Indexed
Not Allows duplicates
Heterogeneous data allowed (Multiple data types)

```python
# Empty FrozenSet
empty_frozenset = frozenset()
print(empty_frozenset)

# Integer FrozenSet
numbers = frozenset([1, 2, 3, 4, 5])
print(numbers)

# Mixed Data Type FrozenSet
mixed = frozenset([1, "apple", 3.5, True])
print(mixed)

# Nested FrozenSet
nested = frozenset([
                        frozenset([1,2]),
                        frozenset([3,4])
                    ])
print(nested)
```

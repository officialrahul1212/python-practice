### Dictionary {}

A dictionary  is a collection of key: value pais used to store data

A Dictionary is a collection data type used to store data in the form of:

Key : Value

list store values in []

### Characteristics
- Ordered 
- Mutable (changeable = add , update, delete)
- Indexed using keys 
- Stores data in key-value pairs 
- Duplicate keys not allowed 
- Duplicate values allowed 
- Heterogeneous Data Types Allowed




- Advantages of Dictionary:
Fast searching
Stores labeled data
Mutable
Flexible


- Disadvantages of Dictionary:
Duplicate keys not allowed
More memory usage
Keys must be unique

### 🔘Create Dictionary

```python
dict0={}              # empty dictionary
print("empty dictionary =",dict0)

dict1={
    "name":"rahul",
    "roll no":18
}
print(dict1)

dict2={
    "student1":{
        "name":"abc",
        "age":23
    },
    "student2":{
        "name":"xyz",
        "age":23
    }
}
print("nested dictionary=",dict2)



#using dict constructor creacte dictionary
stu= dict(name="ggg", age=21)
print("using dict constructor= ",stu)

```

### 🔶Dictionary Methods


```python
a={
    "name":"rahul",
    "age":23
}


#1️⃣.key() -Return all keys
print(a.keys())

#2️⃣.values() -Rerturn all values
print(a.values())

#3️⃣.items() - Return key value pair
print(a.items())

#4️⃣.update() - update existing values and add new key
print(a.update({"age":30}))                            #update age
print(a)
print(a.update({"course":"data analytics"}))             #add new key and value
print(a)

#5️⃣.pop()  - remove specific key
print(a.pop("course"))
print(a)

#6️⃣.popitem()  - remove last inserted element
a.popitem()
print(a)

#7️⃣.clear() - remove all items
a.clear()
print(a)

#8️⃣.copy() - create duplicate dictionary
b={"age":22}
c=b.copy()
print(c)

#9️⃣setdefault() - add key only if it does not exist
b.setdefault("year",2026)
print(b)

#🔟add new key
b["city"]="sangli"
print(b)
```

### 🔷Accessing Elements

 - get()         --> Returns value using key safely (return none if key is missing)
 - keys()        --> Returns all keys
 - values()      --> Returns all values
 - items()       --> Returns key-value pairs as tuples

```python
student_details = {
                    "name":"Rahul",
                    "age":22,
                    "course":"Python"
                  }

print(student_details.get("name"))            #return none if key is missing

print(student_details["name"])                 #return error if key is missing

print(student_details.keys())

print(student_details.values())

print(student_details.items())
```

### 🔶Dictionary Built-in Functions

- len()         --> Returns total key-value pairs
- max()         --> Returns largest key
- min()         --> Returns smallest key
- sum()         --> Returns sum of numeric keys
- sorted()      --> Sorts dictionary keys

```python
student_marks = {
                    1:85,
                    2:90,
                    3:78
                 }

print(len(student_marks))

print(max(student_marks))

print(min(student_marks))
```

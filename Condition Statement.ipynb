# Conditional Statements

👉 Conditional statements are used to make decisions in Python.

👉 They execute different blocks of code based on conditions.

## types

1) if statement
2) if-else statement
3) if-elif-else statement
4) Nested if
5) Short-hand if (Ternary Operator)

## 1. if Statement

👉 Executes code only if condition is True

Syntax

if condition:
    statement

```python
age = 18

if age >= 18:
    print("Eligible for voting")
```

## 2. if-else Statement

👉 Executes one block if condition is true, otherwise another block.

Syntax

if condition:
    statement
else:
    statement

```python
num = 5

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

## 3. if-elif-else Statement

👉 Used to check multiple conditions.

Syntax


if condition1:
    statement
    
elif condition2:
    statement
    
else:
    statement

```python
marks = 55

if marks >= 90:
    print("Grade A")
    
elif marks >= 70:
    print("Grade B")
    
elif marks>=50:
    print("Grade C")
    
else:
    print("Grade D")
```

## 4. Nested if Statement

👉 if inside another if

Syntax


if condition1:
    # Executes if condition1 is True
    
    if condition2:
        # Executes if both conditions are True

```python
# username = "admin"
# password = "12345"

username = str(input("Enter username: "))
password = str(input("Enter password: "))

if username == "admin":
    
    if password == "1234":
        print("Login Successful")
    else:
        print("condition 2 false")

else:
        print("condition 1 false") 
```

## 5. Short-hand if (Ternary Operator)

👉 One-line condition

Syntax

value_if_true if condition else value_if_false

```python
age = 20

result = "Adult" if age >= 18 else "Minor"

print(result)
```

## Logical Operators in Conditional statement

### And 
both condition is true then work

### Or
one condition is true then work 

```python
age = 12
has_id = True

if age >= 18 and has_id:          # false  and   true 
    print("Entry Allowed")
else:
    print("not allowed")
```

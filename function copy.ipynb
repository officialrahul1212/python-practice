```python
def student_details():

    student_name = "Rahul"

    print(student_name)

student_details()
```

```python
course_name = "Python"

def display_course():

    print("inside fun--",course_name)

display_course()

print("outside func--",course_name)
```

### 1. Positional Parameters & Positional Arguments

```python
# Function Definition

def student_details(student_name, student_age):

    # student_name and student_age are Positional Parameters

    print("Name :", student_name)

    print("Age  :", student_age)


# Function Call

student_details("Rahul", 22)

# "Rahul" and 22 are Positional Arguments
```

```python
def student_details(student_name, student_age):

    print("Name :", student_name)

    print("Age  :", student_age)


student_details(22, "Rahul")
```

### 2. Default Parameters & Default Arguments

```python
# Without Passing Argument

def greet(student_name="Guest"):

    print("Hello", student_name)


# No Argument Passed

greet()
```

```python
# Passing Argument

def greet(student_name="Guest"):

    print("Hello", student_name)


# Argument Passed

greet("Rahul")
```

### 3. Keyword Parameters and Keyword Arguments

```python
def student_details(student_name, student_age):
                # Parameters: student_name , student_age
    print("Name :", student_name)
    print("Age  :", student_age)


student_details(
                student_age=22,
                student_name="Rahul"
               )

#Arguments:  student_age=22 , student_name="Rahul"
```

# Function With return and Without return 

## 1. Function Without Return


A function that does not use the return statement is called a Function Without Return.
It only performs a task and displays the result. can't store result easily.

```python
def add_numbers():

    number1 = 10
    number2 = 20

    print( number1 + number2)


add_numbers()
```

```python
def add_numbers():

    number1 = 10
    number2 = 20

    print( number1 + number2)

result = add_numbers()

print("result =",result)
```

## 2. Function With Return


A function that sends a value back to the caller using the return statement is called a Function With Return.

```python
def add_numbers():

    number1 = 10

    number2 = 20

    return number1 + number2


add_numbers()
```

```python
def add_numbers():

    number1 = 10
    number2 = 20

    #print( number1 + number2)

    return number1 + number2
    
result = add_numbers()

print("result =",result)
```

# map(), filter(), and reduce() in Python

## 1) map()

```python
#Without map()

numbers = [1, 2, 3, 4, 5]

squares = []

for i in numbers:

    squares.append(i ** 2)

print(squares)
```

```python
#With map()

numbers = [1, 2, 3, 4, 5]

result = map(
             lambda i : i ** 2,numbers
            )

print(list(result))
```

```python
student_names = [
                 "rahul",
                 "amit",
                 "priya"
                ]

result = map(
             lambda i : i.upper(), student_names
            )

print(list(result))
```

```python
list1 = [10, 20, 30]

list2 = [1, 2, 3]

result = map(
             lambda i, j : i + j,list1,list2
            )

print(list(result))
```

```python
numbers = [1, 2, 3]

result = map(
             lambda i : i ** 2,numbers
            )

print(result)
```

```python
numbers = [1, 2, 3]

result = map(
             lambda number : number ** 2,
             numbers
            )

print(list(result))
```

## 2) filter()

```python
# Without filter()

numbers = [1, 2, 3, 4, 5, 6]

even_numbers = []

for number in numbers:

    if number % 2 == 0:

        even_numbers.append(number)

print(even_numbers)
```

```python
#With filter()

numbers = [1, 2, 3, 4, 5, 6]

result = filter(
                lambda number : number % 2 == 0,
                numbers
               )

print(list(result))
```

```python
numbers = [1, 2, 3, 4, 5, 6]

result = filter(
                lambda number : number % 2 != 0,
                numbers
               )

print(list(result))
```

```python
student_marks = [35, 80, 25, 90, 45]

result = filter(
                lambda mark : mark >= 40,
                student_marks
               )

print(list(result))
```

```python
numbers = [1, 2, 3]

result = filter(
                lambda number :
                number > 1,
                numbers
               )

print(result)
```

```python
numbers = [1, 2, 3]

result = filter(
                lambda number :
                number > 1,
                numbers
               )

print(list(result))
```

## 3) reduce()

```python
#Without reduce()

numbers = [10, 20, 30, 40]

total = 0

for number in numbers:

    total += number

print(total)
```

```python
#With reduce()

from functools import reduce

numbers = [10, 20, 30, 40]

result = reduce(
                lambda number1, number2 :
                number1 + number2,
                numbers
               )

print(result)
```

```python
from functools import reduce

numbers = [45, 78, 22, 99, 34]

result = reduce(
                lambda number1, number2 :
                number1 if number1 > number2 else number2,
                numbers
               )

print(result)
```
